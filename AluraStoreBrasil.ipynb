# Importação dos dados:



```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
def formatador_reais(valor, pos):
  return f'R${valor:,.0f}'
formatador = FuncFormatter(formatador_reais)

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
faturamento1 = loja['Preço'].sum()
faturamento2 = loja2['Preço'].sum()
faturamento3 = loja3['Preço'].sum()
faturamento4 = loja4['Preço'].sum()
dados_faturamento = pd.Series([faturamento1, faturamento2, faturamento3, faturamento4], index = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'])
```

```python

print(f'Faturamento da Loja 1: {faturamento1}')
print(f'Faturamento da Loja 2: {faturamento2}')
print(f'Faturamento da Loja 3: {faturamento3}')
print(f'Faturamento da Loja 4: {faturamento4}')
```

# 2. Vendas por Categoria

```python

cat_mais_vendido1 = loja.groupby('Categoria do Produto').size()
cat_mais_vendido2 = loja2.groupby('Categoria do Produto').size()
cat_mais_vendido3 = loja3.groupby('Categoria do Produto').size()
cat_mais_vendido4 = loja4.groupby('Categoria do Produto').size()
categorias_dados = pd.DataFrame({'Loja 1': cat_mais_vendido1, 'Loja 2': cat_mais_vendido2,'Loja 3': cat_mais_vendido3,'Loja 4': cat_mais_vendido4})
```

```python
print(f'Categoria de produto mais vendida na loja 1: \n{cat_mais_vendido1.to_string()}')
print(f'\nCategoria de produto mais vendida na loja 2: \n{cat_mais_vendido2.to_string()}')
print(f'\nCategoria de produto mais vendida na loja 3: \n{cat_mais_vendido3.to_string()}')
print(f'\nCategoria de produto mais vendida na loja 4: \n{cat_mais_vendido4.to_string()}')
```

# 3. Média de Avaliação das Lojas


```python
media_loja1 = loja['Avaliação da compra'].mean()
media_loja2 = loja2['Avaliação da compra'].mean()
media_loja3 = loja3['Avaliação da compra'].mean()
media_loja4 = loja4['Avaliação da compra'].mean()
dados_avaliacoes = pd.Series([media_loja1, media_loja2, media_loja3, media_loja4], index=['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'])
```

```python
print(f'Média de avaliações da Loja 1: {media_loja1}')
print(f'Média de avaliações da Loja 2: {media_loja2}')
print(f'Média de avaliações da Loja 3: {media_loja3}')
print(f'Média de avaliações da Loja 4: {media_loja4}')
```

# 4. Produtos Mais e Menos Vendidos

```python
mais_vendido1 = loja.groupby('Produto').size().sort_values(ascending=False).head(5)
menos_vendido1 = loja.groupby('Produto').size().sort_values(ascending=False).tail(5)

mais_vendido2 = loja2.groupby('Produto').size().sort_values(ascending=False).head(5)
menos_vendido2 = loja2.groupby('Produto').size().sort_values(ascending=False).tail(5)

mais_vendido3 = loja3.groupby('Produto').size().sort_values(ascending=False).head(5)
menos_vendido3 = loja3.groupby('Produto').size().sort_values(ascending=False).tail(5)

mais_vendido4 = loja4.groupby('Produto').size().sort_values(ascending=False).head(5)
menos_vendido4 = loja4.groupby('Produto').size().sort_values(ascending=False).tail(5)
```

```python
print(f'Produtos mais vendidos Loja 1: \n {mais_vendido1.to_string()}')
print(f'\nProdutos menos vendidos Loja 1: \n {menos_vendido1.to_string()}')
print('_________________________________')
print(f'\nProdutos mais vendidos Loja 2: \n {mais_vendido2.to_string()}')
print(f'\nProdutos menos vendidos Loja 2: \n {menos_vendido2.to_string()}')
print('_________________________________')
print(f'\nProdutos mais vendidos Loja 3: \n {mais_vendido3.to_string()}')
print(f'\nProdutos menos vendidos Loja 3: \n {menos_vendido3.to_string()}')
print('_________________________________')
print(f'\nProdutos mais vendidos Loja 4: \n {mais_vendido4.to_string()}')
print(f'\nProdutos menos vendidos Loja 4: \n {menos_vendido4.to_string()}')
```

# 5. Frete Médio por Loja

```python
frete_media_loja1 = loja['Frete'].mean()
frete_media_loja2 = loja2['Frete'].mean()
frete_media_loja3 = loja3['Frete'].mean()
frete_media_loja4 = loja4['Frete'].mean()
dados_frete = pd.Series([frete_media_loja1, frete_media_loja2, frete_media_loja3, frete_media_loja4], index = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'])
```

```python
print(f'Média de frete da Loja 1: {frete_media_loja1}')
print(f'Média de frete da Loja 2: {frete_media_loja2}')
print(f'Média de frete da Loja 3: {frete_media_loja3}')
print(f'Média de frete da Loja 4: {frete_media_loja4}')
```

# Visualização (Gráficos)


Faturamento


```python
grf1 = dados_faturamento.plot(kind = 'bar', title = 'Faturamento total por loja', rot = 0, figsize = (10, 6))
grf1.yaxis.set_major_formatter(formatador)
```

Vendas por categoria

```python
plt.figure(figsize =(10,6))
plt.stem(dados_frete.index, dados_frete.values)
plt.title('Frete Médio por Loja')
plt.ylabel('Custo Médio do Frete (R$)')
plt.xlabel('Loja')
plt.gca().yaxis.set_major_formatter(formatador)
plt.show()
```

categorias


```python
comparacao_categorias = categorias_dados.T

grf_cat = comparacao_categorias.plot(kind='bar',figsize=(15, 7), title='Comparação entre Categorias em cada Loja', rot=0)
grf_cat.legend(title='Categorias de Produtos', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.ylabel('Quantidade Vendida')
plt.xlabel('Lojas')
for barra in grf_cat.patches: grf_cat.annotate(f'{barra.get_height():.0f}', (barra.get_x() + barra.get_width() / 2, barra.get_height()), ha='center', va='bottom', xytext=(0, 5), textcoords='offset points')
```

médias

```python
plt.figure(figsize=(10, 6))
plt.scatter(y=dados_avaliacoes.values, x=dados_avaliacoes.index, s=150)
plt.title('Média de Avaliação por Loja')
plt.xlabel('Nota Média (1-5)')
plt.ylabel('Loja')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.ylim(3.5, 4.5)
for i, (loja_nome, media_valor) in enumerate(dados_avaliacoes.items()):
    plt.text(y=media_valor + 0.05, x=i, s=f'{media_valor:.2f}', va='center',fontsize=12)
```
